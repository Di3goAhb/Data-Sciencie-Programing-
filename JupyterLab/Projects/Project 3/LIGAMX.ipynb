Tabla de diferencia de goles y goles totales para determinar el porcentaje de efectividad de un equipo.

```python
plt.scatter(liga['GOLESFAVOR'], liga['TIROS_A_GOL'])

plt.xlabel('DIFGOLES')
plt.ylabel('TIROS_A_GOL')

plt.show()
```

Por ejemplo: para el america los tiros a gol fueron de 132 y sus goles totales 44 = 33% de tiros a gol entran a la porteria

AquÃ­ vemos una tabla de goles por el equipo el cual podemos observar es similar o sigue una tendencia que a mayor posiciÃ³n de la tabla mayor cantidad de goles

```python
equipos = liga['EQUIPO']
difgoles = liga['DIFGOLES']


plt.bar(equipos, difgoles)

plt.xlabel('EQUIPO')
plt.ylabel('GOLES')

plt.xticks(rotation=90)

plt.show()
```

Observamos el valor del equipo en comparaciÃ³n con su posiciÃ³n en la table e igual vemos una cierta tendencia que a mayor valor del equipo, mayor posiciÃ³n en la tabla.

```python
plt.plot(liga['EQUIPO'], liga['VALOR'])

plt.xlabel('EQUIPOS')
plt.ylabel('VALOR')

plt.xticks(rotation=90)

plt.show()
```

```python
datos_variable1 = liga['PGANADOS']
datos_variable2 = liga['PEMPATADOS']
datos_variable3 = liga['PPERDIDOS']

plt.figure(figsize=(8, 6))

sns.histplot(datos_variable1, bins=5, kde=True, label='Partidos ganados')


sns.histplot(datos_variable2, bins=5, kde=True, label='Empates')


#sns.histplot(datos_variable3, bins=5, kde=True, label='Partidos perdidos')



plt.xlabel('Partidos ganados/empatados/perdidos')
plt.ylabel('Num. de equipos')
plt.title('Histogramas de partidos')

plt.legend()
plt.show()
```

```python
datos_variable1 = liga['GOLESFAVOR']
datos_variable2 = liga['GOLESCONTRA']
datos_variable3 = liga['DIFGOLES']

plt.figure(figsize=(8, 6))

sns.histplot(datos_variable1, bins=5, kde=True, label='Goles a favor')

sns.histplot(datos_variable2, bins=5, kde=True, label='Goles en contra')

sns.histplot(datos_variable3, bins=5, kde=True, label='Dif de goles')


plt.xlabel('goles favor/encontra')
plt.ylabel('Numero de equipos')
plt.title('Histogramas de golees')

plt.legend()
plt.show()
```

Histograma de los ultimos partidos de cada equipo para observar como se comportaron los equipos al finalizar el torneo. Hay mÃ¡s equipos con victorios que derrotas antes de acabar

```python
datos_variable1 = liga['LAST_5_WIN']
datos_variable2 = liga['LAST_5_EMP']
datos_variable3 = liga['LAST_5_LOST']

plt.figure(figsize=(8, 6))

sns.histplot(datos_variable1, bins=5, kde=True, label='Ultimas 5 victorias')

#sns.histplot(datos_variable2, bins=5, kde=True, label='Ultimos 5 empates')

#sns.histplot(datos_variable3, bins=5, kde=True, label='Ultimas 5 derrotas')

plt.xlabel('victorias/derrotas/empates')
plt.ylabel('Numero de equipos')
plt.title('Histogramas de ultimos partidos')

plt.legend()
plt.show()
```
